# **7. Split the dataset into features and labels. Use your intuition to determine which column indicates the labels.**

```python
from sklearn.model_selection import train_test_split
```

```python
y=income['income_>50K'].copy()
x=pd.DataFrame(income.drop(['income_>50K'],axis=1))
x.head(10)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state=0)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

# ***income_>50K column is my label for Income Dataset(50k).csv***

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(X_train, y_train)

print("Test set accuracy: {:.2f}".format(knn.score(X_test, y_test)))

"""**After Scaling**"""
Y= income_scaled['income_>50K'].copy()
X=pd.DataFrame(income_scaled.drop(['income_>50K'],axis=1))
```
