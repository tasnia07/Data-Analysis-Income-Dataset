# **2. Perform classification and calculate accuracy using logistic regression (Use sklearn library). Use 8:2 train-test split.**

```python
# Import the dependencies for logistic regression
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
#Split the data into 80% training and 20% testing
X_train, X_test, y_train, y_test = train_test_split(x, y,test_size=0.2, random_state=42)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
#Train the model
model = LogisticRegression()
model.fit(X_train, y_train) #Training the model
predictions = model.predict(X_test)
print(predictions)# printing predictions
```

```python
#logistic regression

logistic_accuracy=accuracy_score(y_test, predictions)
print( logistic_accuracy)
```

# **3. Perform classification and calculate accuracy using decision tree (Use sklearn library). Use 8:2 train-test split.**

```python
#Decision tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
X = income.iloc[:,0:6]  #considering income_>50K as the label
y = income.iloc[:, 6]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
clf = DecisionTreeClassifier(criterion='entropy',random_state=1)
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
score=accuracy_score(y_pred,y_test)
print(score)
```

# **4. Compare the accuracy and plot them as a bar chart using matplotlib/seaborn**

```python
fig =plt.figure()
ax=fig.add_axes([0,0,0.5,1.5])
approach = ['Logistic Regression', 'Decision Tree']
percentage = [logistic_accuracy, score]
ax.bar(approach,percentage)
ax.set_title('Accuracy comparison')
plt.ylabel('Percentage')

print("Accuracy using logistic regression :", logistic_accuracy)
print("Accuracy using decision tree :", score )
print("Comparison is shown in the graph below :")
```
